The following cell downloads the Africa Galore dataset, stores it in a Pandas dataframe, and prints some basic statistics.

```python
# Download the file from a link.
filepath = "https://storage.googleapis.com/dm-educational/assets/ai_foundations/africa_galore_large.json"

# Read with pandas.
df = pd.read_json(filepath, lines=True)

# Other file-type examples:
# For loading JSON files.
# df = pd.read_json(filepath, orient="records")

# For loading a CSV file.
# df = pd.read_csv(filepath, encoding="utf-8")

# If you are using an API, implement a function, such as `load_from_api()`
# that downloads the data using an API.
# df = load_from_api()

# Take a quick look at the first few rows.
display(df.head())

# Take a detailed look at the first example:
print("-" * 80 + "\nFirst example:\n")

print(f"Category: {df['category'][0]}")
print(f"Name: {df['name'][0]}")
print(f"Description: {textwrap.fill(df['description'][0])}")

print("\n" + "-" * 80 + "\n")

print("Columns:", df.columns.tolist())
print("Number of rows:", len(df))

# Display unique values from the 'category' column.
print("Unique categories:", df["category"].unique())
```

## Compute statistics of your dataset

To get a better sense of the composition of your dataset, you can compute some summary statistics of your data.

The following code gives you a breakdown of how many examples exist per category.

```python
# Compute the number of examples per category.
category_counts = Counter(df["category"])

print("Number of examples per category:\n")
for cat in category_counts:
  print(f"{cat}: {category_counts[cat]}")
```

As you can see, this dataset has many more examples with the category "Story" than examples with other categories. If this is also true for your dataset, you likely will have to balance the number of examples across categories.
